### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
dataTimeFly
```

```python
#conversió de milles a km
Distance_km = dataTimeFly['Distance']*1.60934
Distance_km
```

```python
dataTimeFly_hour = dataTimeFly['AirTime']/60
dataTimeFly_hour
```

```python
AvergSpeed = round(Distance_km/dataTimeFly_hour,2)
AvergSpeed
```

```python
dataDelay.assign(AvergSpeed=AvergSpeed)
```
